```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

from IPython.display import SVG
```

## The dot motion paradigm

An experimental participant is presented with the following visual stimulus, except that all the dots are the same colour. They see about 20 dots moving in random directions, except for two of these which have "motion coherence," which means that they are moving in the same direction. So the movement is random, except for the 10% which are coherent. The participant does not know which direction is the coherent one. Their task is to give a response (e.g., press a button) when they are able to perceive the coherent motion.

![dot motion paradigm](./dotmotion.svg)

## Normal distribution plot

The plot below represents the perceptual evidence associated with the stimuli. Perceptual processes are noisy, which means that they are prone to error. 

```python
# Generate the distribution

mean = 0.1
variance = 1
sd = math.sqrt(variance)

x = np.linspace(mean - 3*sd, mean + 3*sd, 100)
plt.plot(x, stats.norm.pdf(x, mean, sd))

# Style the plot

plt.grid(b=True)
plt.style.use('fast')

plt.ylabel('Density')
plt.xlabel('Sensorimotor Parameter')

plt.savefig('normaldist.png', dpi=500)
plt.show()
```
