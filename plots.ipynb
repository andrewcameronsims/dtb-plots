```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

## Normal distribution plot

```python
mu = 0.1
variance = 1
sigma = math.sqrt(variance)

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))

# Styles

plt.grid(b=True)
plt.style.use('fast')


plt.show()
```
